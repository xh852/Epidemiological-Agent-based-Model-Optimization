```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
opt_results = pd.read_csv('compile_opt_results.csv')
```

```python
opt_results.iloc[0][1:5]
```

```python
# Plot change in seconds
width = 0.2
ind = np.arange(4) 
for i in range(3):
    plt.bar(ind+(width*i),opt_results.iloc[i][1:5],width,label=opt_results.iloc[i][0])
plt.xlabel("Optimization Method")
plt.xticks(ind+width, opt_results.columns[1:5], rotation=10, ha='center')
plt.ylabel("Seconds", rotation=0, labelpad=25)
plt.title("Agent-based Simulation Speed")
plt.legend()
plt.show()
```

```python
# Plot change in seconds
for i in range(3):
    plt.plot(opt_results.iloc[i][5:],"o", label=opt_results.iloc[i][0])
plt.xlabel("Optimization Method")
plt.xticks([0,1,2,3], opt_results.columns[1:5], rotation=10, ha='center')
plt.ylabel("# Times\nfaster vs.\nOriginal", rotation=0, labelpad=30)
plt.title("Agent-based Simulation Speed")
plt.legend()
plt.show()

width = 0.2
ind = np.arange(4) 
for i in range(3):
    plt.bar(ind+(width*i),opt_results.iloc[i][5:],width,label=opt_results.iloc[i][0])
plt.xlabel("Optimization Method")
plt.xticks(ind+width, opt_results.columns[1:5], rotation=10, ha='center')
plt.ylabel("# Times\nfaster vs.\nOriginal", rotation=0, labelpad=30)
plt.title("Agent-based Simulation Speed")
plt.legend()
plt.show()
```
